# Determining the Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
col_list = ["Candidate Ref", "Duration to accept offer", "Notice period", "Percent difference CTC","Status" ]
```

```python
for i in col_list:
    data[i] = le.fit_transform(data[i])
```

```python
clf=DecisionTreeClassifier(min_samples_split=1800)
```

```python
clf=clf.fit(x,y)
```

```python
from sklearn import tree
```

```python
tree.plot_tree(clf)
```
