# Now lets start creating the model:
![](https://i.imgflip.com/328qto.gif)

````python
# Sperating data into X and Y
X = df['Total']
X.head()
````

````python
# Sperating data into X and Y
y = df['Rating']
y.head()
````

````python
# Import train-test scikit learn
from sklearn.model_selection import train_test_split
````

````python
# Split the data for train and test 
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.75,random_state=100)
````

````python
#Since x_train is a series without a column so we only get a sigle value. Meaning that X_train has 750 rows(since we have chosen
#our training size as 75%) and 0 columns.

X_train.shape
````

````python
X_train.values.reshape(-1,1)
````

````python
X_train.values.reshape(-1,1).shape
````

````python
#Overwriting X_train and X_test after adding a column.
X_train = X_train.values.reshape(-1,1)
X_test = X_test.values.reshape(-1,1)
````

# Time to fit the model:

````python
# Importing Linear Regression model from scikit learn
from sklearn.linear_model import LinearRegression
````

````python
# Initializing.
lr = LinearRegression()
````

````python
# Fitting the model:
lr.fit(X_train,y_train)
````

````python
lr.predict([[1050.0]])
````

````python
# Predicting the ratings for the Test values
y_pred_test = lr.predict(X_test)
````

````python
y_pred_test
````

````python
y_test
````

# Checking the test accuracy.

````python
# Importing metrics for the evaluation of the model
from sklearn.metrics import r2_score,mean_squared_error
````

````python
# calculate Mean square error
mean_squared_error(y_test,y_pred_test)
````

````python
mse = mean_squared_error(y_test,y_pred_test)
````

````python
# Calculate R square vale
r2_score(y_test,y_pred_test)
````

````python
x_range = [i for i in range(len(y_test))]
````

````python
plt.xlabel('Total')
plt.ylabel('Ratings')
plt.title('Prediction')

plt.plot(x_range,y_test)
plt.plot(x_range,y_pred_test)

# Function add a legend  
plt.legend(["Real", "Predicted"])
````

````python
plt.show()
````

````python
# Create a list of values in the best fit line
abline_values = [lr.coef_ * i + lr.intercept_ for i in range(0,15,1)]
x = [i for i in range(0,15,1)]

# Plot the best fit line over the actual values

plt.scatter(df['Total'],df['Rating'])
plt.plot(x, abline_values, 'b')

plt.title('Our Fitted Line')
plt.show()
````

# Checking the train accuracy.
