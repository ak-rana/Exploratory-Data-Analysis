# Determining the decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
col_list = ["Age", "Gender", "Education Level", "Job Title"]
```

```python
for i in col_list:
    data[i] = le.fit_transform(data[i])
```

```python
data
```

```python
data
```

```python
x =  data[["Age", "Gender", "Education Level", "Job Title", "Years of Experience"]]
```

```python
y = data["Salary"]
```

```python
x
```

```python
y
```

```python
clf=DecisionTreeRegressor(min_samples_split=100)
```

```python
clf= clf.fit(x,y)
```

```python
from sklearn import tree
```

```python
tree.plot_tree(clf)
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.figure(figsize = (11,11))
tree.plot_tree(clf, fontsize=10)
plt.show()
```

# Creating the model

```python
X = data['Years of Experience']
X.head()
```

```python
y = data['Salary']
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.70,random_state=100)
```

```python
X_train.shape
```

```python
X_train.values.reshape(-1,1)
```

```python
X_train.values.reshape(-1,1).shape
```

```python
X_train = X_train.values.reshape(-1,1)
X_test = X_test.values.reshape(-1,1)
```

# Time to fit the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.predict([[1050.0]])
```

```python
y_pred_test = lr.predict(X_test)
```

```python
y_pred_test
```

```python
y_test
```

# Checking the test accuracy

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
# calculate Mean square error
mean_squared_error(y_test,y_pred_test)
```

```python
mse = mean_squared_error(y_test,y_pred_test)
```

```python
r2_score(y_test,y_pred_test)
```

```python
x_range = [i for i in range(len(y_test))]
```

```python
plt.xlabel('Job Title')
plt.ylabel('Salary')
plt.title('Prediction')

plt.plot(x_range,y_test)
plt.plot(x_range,y_pred_test)

# Function add a legend  
plt.legend(["Real", "Predicted"])
```
